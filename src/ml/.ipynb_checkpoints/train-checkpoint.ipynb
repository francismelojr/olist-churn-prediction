```python
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)

import sqlite3
import seaborn as sns
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
# Importing Table

path = r'C:\Users\franc\Data Science\Notebooks\o-list\data\olist.db'
with sqlite3.connect(path) as conn:
    query = 'SELECT * FROM abt_olist_churn'
    df = pd.read_sql_query(query, conn)
```

```python
df.head()
```

```python
# out of time sample

df_oot = df[df['dtReference'] == '2018-02-01']

df_train = df[df['dtReference'] != '2018-02-01']
```

```python
df_train.shape, df_oot.shape
```

```python
var_identity = ['dtReference', 'idVendedor']
target = ['flChurn']

features = df.columns.tolist()
features = list(set(features) - set(var_identity + target))
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    df_train[features], df_train[target], test_size=0.2, random_state=23)
```
